```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 60, 1000)
```

Power generated by Mark I : t

```python
p_1 = t
```

Power generated by Mark II : 4 * (t - 30)

```python
p_2 = 4 * (t - 30)
```

```python
fig, ax = plt.subplots()
plt.title('The Race of Power')
plt.xlabel('Time (in days)')
plt.ylabel('Power Generated (in kJ)')
ax.set_xlim([0, 60])
ax.set_ylim([0, 100])
ax.plot(t, p_1, c='green')
ax.plot(t, p_2, c='red')
plt.axvline(x=40, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```
